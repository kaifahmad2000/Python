23. What are the challenges associated with memory management in Python?
- Memory management in python is largely handled by interpreter but several challenges still arise like: Memory leaks, high memory usage, mutable and immutable objects, GIL.

i) Memory leaks: Despite garbage collection, Python can still suffer from memory leaks due to lingering refrences, circular references and third party libraries.

ii) High Memory Usage: Python stores metadeta with objects, making them heavier. Lists, dictionaries and custom objects may consumes more memory due to dynamic resizing and overhead.

iii) Mutable and Immutable Objects: Mutability can lead to unexpected memory behaviour if objects are shared unintentionally like: modifying the mutable objects ie., lists can affects all references to it.

iv) GIL: GIL is Global Interpreter Lock, it prevents true parallel execution in python which can indirectly affect memory use in concurrent programs.

```python
#6. Write a Python program that logs an error message to a log file when a division by zero exception occurs.
import logging
logging.basicConfig(filename="error_log.log",
                    level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')
def divide(num1, num2):
  try:
    return num1/num2
  except ZeroDivisionError as e:
    logging.error("Cannot divide a number by zero,",e)
    print("Error cannot divide a number by zero")
```

```python
#14. Write a program that uses the logging module to log both informational and error messages.
import logging
logging.basicConfig(filename = "test.log",
                    level = logging.DEBUG,
                    format = '%(asctime)s - %(levelname)s - %(message)s')
l = [1, 'hello',[2, 'world'], 4, [3, 'Python']]
l1_int = []
l2_str = []

for i in l:
    logging.info(f"Processing each element: {i}")

    if type(i) == list:
        for j in i:
            logging.info(f"Processing sublist element: {j}")
            if type(j) == int:
                l1_int.append(j)
            else:
              l2_str.append(j)
    elif type(i) == int:
        l1_int.append(i)
    else:
        l2_str.append(i)

logging.info(f"The result is: {l1_int}, {l2_str}")
logging.shutdown()
```

```python
#18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

log_handler = RotatingFileHandler(
    filename = 'rotate_log',
    maxBytes = 1 * 1024 * 1024,
    backupCount = 3
)

formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
log_handler.setFormatter(formatter)

logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.addHandler(log_handler)

for i in range(1000):
  logger.info(f"This is log entry {i}")
```

```python
#21. Write a Python program that reads a file and prints the number of occurrences of a specific word.
def count_word_occurences(filename, word):
  try:
    with open(filename, 'r') as file:
      content = file.read()
      content = content.lower()
      word = word.lower()
      count = content.split().count(word)
      print(f"The word {word} occurs {count} times in the file.")
  except FileNotFoundError as e:
      print("Error occured while searching for file,", e)
count_word_occurences('file.txt', 'line')
```
