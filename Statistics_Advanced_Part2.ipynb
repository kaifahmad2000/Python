1. What is hypothesis testing in statistics?
- Hypothesis testing is a formal method used to make inferences about a population based on a sample.
It allows us to test assumptions about a population parameter.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The default assumption or the claim we try to test against is known as null hypothesis while an alternate hypothesis is the claim you're testing for.
The main difference is in the assumptions as null hypothesis default or status quo while alternate hypothesis is what are we going to prove.

3. What is the significance level in hypothesis testing, and why is it important?
- The significance level, denoted by alpla, is a threshold that defines how much evidence you require before rejecting the null hypothesis.
It is important because it controls Type-I error, guides in decision making and sets the bar for evidence.

4. What does a P-value represent in hypothesis testing?
- The p-value is the probability value that is calculated from a statistical test corresponding to the Z-score of the data point.
p-value in hypothesis testing is used to decide whether to reject a null hypothesis or not.

5. How do you interpret the P-value in hypothesis testing?
- The p-value helps in evaluating the strength of evidence against the null hypothesis.e.g., we have null hupothesis: A medicine has no effect.
test result in p-value: 0.03
Significance level: 0.05
So, we can interpret that there is only 3% chance that the medicine has no effect which is less than significance level.
Therefore, we reject the null hypothesis and conclude that medicine likely has an effect.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type-I error is the rejection of null hypothesis when it is actually true.
e.g., the conviction of an innocent.

Type-II error is the failure to reject the null hypothesis when it is actually false. e.g., a guilty person may not be convicted.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- When the parameter is greater than or less than a certain value we go for a one tailed test while if the alternate hypothesis as not equality symbol we go for two tailed test.

8. What is the Z-test, and when is it used in hypothesis testing?
- Z-test is a type of statistical method used for hypothesis testing which deals with the average value of the population.
There are certain conditions to used Z-test:
i) Sample size should be greater than 30.
ii) stadard deviation of the population should be known.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- Z = (sample mean - population mean)/ (population standard deiation/ square root of sample size)
* It standardize the result to the standard normal distribution.
* Helps calculate a p-value which certainly helps in hypothesis conclusion.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T- distribution is a probability distribution that is:
    -symmetrical and bell-shaped, like the normal distribution.
    -flatter and wider than the normal distribution.
    -has heavier tails, meaning it accounts better for extreme values.
T-distribution is used instead of normal distribution when population standard deviation is unknown, sample size is small (less than 30) and data is approx normally distributed or symmetric.

11. What is the difference between a Z-test and a T-test?
- i) In Z-test population SD is known while unknown in T-test.
ii) Sample size should be greater than or equal to 30 for z- test while t-test could be performed in less than 30.
iii) Z-test uses Standard Normal Distribution while T-test uses          T-distribution.
iv) Z-test has thinner tails while T-test has thicker tail.

12. What is the T-test, and how is it used in hypothesis testing?

  - a sample mean and a known value.
  - two sample means.

T-test is used in hypothesis testing when the sample size < 30 and the sample standard deviation is given.

C.I = Point Estimate =- Margin of error

15. What is the margin of error, and how does it affect the confidence interval?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, ttest_ind, ttest_rel, ttest_1samp, t, chisquare, chi2_contingency

import warnings
warnings.filterwarnings('ignore')
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

# sample data of weight of the individuals.

np.random.seed(0)
sample_size = 30
sample_data = np.random.normal(loc = 72, scale = 8, size = sample_size)

population_mean = 70
population_std = 8
n = sample_size

sample_mean = np.mean(sample_data)

z_score = (sample_mean-population_mean)/ (population_std/ np.sqrt(n))

p_value = (1 - norm.cdf(abs(z_score)))   # two tailed test

print("Sample Mean: ", round(sample_mean,2))
print("Z-score: ", round(z_score, 3))
print("p-value: ", round(p_value, 4))

#Conclusion:
if p_value < 0.025:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
np.random.seed(11)
sample = np.random.normal(loc = 102, scale = 15, size = 64)

population_mean = 100
population_std = 15

sample_mean = np.mean(sample)
n = len(sample)

z_score = (sample_mean - population_mean)/ (population_std/ np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Sample Mean: ", round(sample_mean, 2))
print("Z-score: ", round(z_score, 3))
print("p-value: ", round(p_value, 4))

#plotting
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

z_critical = norm.ppf(1 - 0.05/2)

plt.figure(figsize = (10, 5))
plt.plot(x, y, label = "Standard Normal Distribution")

plt.fill_between(x, y, where = (x < -z_critical), color = "red", alpha=0.5, label ='Rejection region (left)')
plt.fill_between(x, y, where = (x > z_critical), color = "red", alpha=0.5, label ='Rejection region (right)')

plt.axvline(z_score, color = 'blue', linestyle = '--', linewidth=2, label=f"Z = {z_score:.2f}")

plt.title('Two-Tailed Z-Test with Rejection Regions')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

#Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject H₀: Significant difference detected.")
else:
    print("Fail to reject H₀: No significant difference.")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):

    se = sigma / np.sqrt(n)

    z_alpha = norm.ppf(1 - alpha)
    x_crit = mu0 + z_alpha * se

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    y0 = norm.pdf(x, mu0, se)  # Null distribution
    y1 = norm.pdf(x, mu1, se)  # Alternative distribution

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H₀ Distribution (μ₀)', color='blue')
    plt.plot(x, y1, label='H₁ Distribution (μ₁)', color='green')

    # Type I Error (alpha): area under H₀ beyond critical value
    plt.fill_between(x, y0, where=(x >= x_crit), color='red', alpha=0.4, label='Type I Error (α)')

    # Type II Error (beta): area under H₁ left of critical value
    plt.fill_between(x, y1, where=(x < x_crit), color='orange', alpha=0.4, label='Type II Error (β)')

    plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_type1_type2_errors()
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.

np.random.seed(1)
group_A = np.random.normal(loc=75, scale=10, size=30)  # Mean=75, SD=10
group_B = np.random.normal(loc=80, scale=12, size=30)  # Mean=80, SD=12

t_stat, p_value = ttest_ind(group_A, group_B, equal_var=False)

print("Group A Mean:", round(np.mean(group_A), 2))
print("Group B Mean:", round(np.mean(group_B), 2))
print("T-statistic:", round(t_stat, 3))
print("P-value:", round(p_value, 4))

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two groups.")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(data, confidence=0.95):

    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return round(lower_bound, 2), round(upper_bound, 2)

np.random.seed(49)
sample = np.random.normal(loc=100, scale=10, size=100)
confidence_interval(sample)
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

# Given values
P_Disease = 0.01            # P(A)
P_NoDisease = 1 - P_Disease # P(~A)

P_Pos_given_Disease = 0.99      # P(B|A)
P_Pos_given_NoDisease = 1 - 0.95 # False positive rate = 1 - specificity

numerator = P_Pos_given_Disease * P_Disease
denominator = (P_Pos_given_Disease * P_Disease) + (P_Pos_given_NoDisease * P_NoDisease)
posterior = numerator / denominator
posterior = round(posterior, 4)



print(f"Probability of having disease given a positive test: {posterior*100:.2f}%")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

data = {
    'Yes': [20, 15],   # e.g., 20 males liked the product, 15 females did
    'No': [30, 35]
}
table = pd.DataFrame(data, index=['Male', 'Female'])

print("Contingency Table:")
print(table)

chi2, p, dof, expected = chi2_contingency(table)

print(f"\nChi-Square Statistic: {chi2:.4f}")
print(f"P-Value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:")
print(pd.DataFrame(expected, index=table.index, columns=table.columns))

# Conclusion
alpha = 0.05
if p < alpha:
    print(" Reject H₀: Variables are dependent (associated).")
else:
    print(" Fail to reject H₀: Variables are independent.")
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

observed = [8, 9, 10, 11, 12, 10]

# Expected frequencies if the die is fair
expected = [10, 10, 10, 10, 10, 10]

# goodness of fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("👉 Reject H₀: The observed distribution is significantly different from the expected.")
else:
    print("✅ Fail to reject H₀: No significant difference; observed matches expected.")
```
