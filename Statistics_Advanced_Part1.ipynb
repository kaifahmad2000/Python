1. What is a random variable in probability theory?
- A random variable is a function that assigns a real number to each outcome in a sample space of a random experiment.

2. What are the types of random variables?
- There are two types of random variables:
i) Discrete Random Variable.
ii) Continuous Random Variable.

3. What is the difference between discrete and continuous distributions?
- The main difference between discrete and continuous distribution is that a discrete variable takes a specific and distinct values while a continuous variable takes any value within a range.

4. What are probability distribution functions?
- A probibility distribution functions describe how probabilities are distributed over the values of a random variable. It defines the likelihood of each possible outcome. There are two main types based on the kind of random variable involved:
i) Probability Mass Function(for Discrete Variables).
ii) Probability Density Function(for continuous variables).

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- The commulative distribution function describes the probability upto a certain value, the graph shape is always non-decreasing and ranges from 0 to 1 and applies to both discrete and continuous variables on the other hand the probability distribution function describes the likelihood of a value at a specific point, it forms a bell-shaped curve or skewed shape and it applies only to the continuous distribution.

6. What is a discrete uniform distribution?
- If a random variable X can take on n distinct values, and each has equal probability, then X follows a discrete uniform distribution.
  In simple words, A discrete uniform distribution is a probability distribution in which all possible outcomes are equally likely.

7. What are the key properties of a Bernoulli distribution?
- A Bernoulli distribution models the outcome of a single binary experiment e.g., coin toss, yes/no, success/failure.
The key properties are:
i) Binary Outcomes.
ii) Constant Probability.
iii) Independent Trials.
iv) Complimentary Probability.
v) Expected Value(Mean).
vi) Variance.


8. What is the binomial distribution, and how is it used in probability?
- The binomial distribution models the number of successes in a fixed number of independent and identical Bernoulli trials, each with the same probability of success.
Uses: A/B testing, Quality Control, Epidemiology, Machine Learning.

9. What is the Poisson distribution and where is it applied?
- The Poisson distribution describes the number of times an event occurs in a fixed interval of time, area, distance, or volume - given that the events occur independently and at a constant average rate.
Poisson Distribution is used when counting rare events like: number of earthquake in a region per year, number of mutations in a DNA strand per unit length.

10. What is a continuous uniform distribution?
- The continuous uniform distribution models a situation where all values within a certain range are equally likely to occur. It is the continuous counterpart to the discrete uniform distribution.
Charateristics:
i) All values in [a,b] are equally likely.
ii) Probability of a specific point is 0, but interval probabilities are positive.
iii) Used when no values is favoured over another in a continuous range.

11. What are the characteristics of a normal distribution?
- The characteristics of normal distribution are:
i) It forms symmetrical and bell-shaped curve.
ii) Mean,Median and Mode all occur at the center.
iii) It is perfectly symmetric about the mean.
iv) The spread is determined by the standard deviation.
v) The tails are extended indefinitely in both directions but never touches the x-axis.
vi) About 68% data falls within 1 sigma, 95% within 2 sigma and 99.7% within 3 sigma from the mean.

12. What is the standard normal distribution, and why is it important?
- The standard normal distribution is a special case of the normal distribution where mean is 0 and standard deviation is 1.
It is important because it simplifies calculations,standardization, forms the basis for hypothesis testing and confidence intervals.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem states that if you take a large number of independent, random samples of size "n" from any population(with a finite mean and variance), the sampling distribution of the sample mean will approximate a normal distribution, regardless of the shape of the original population.
    This is critical in statistics because it allows inferences about population parameters using sample data, even when the population distribution is unknown or non-normal.

14. How does the Central Limit Theorem relate to the normal distribution?
- The CLT is closely related to the normal distribution because it states that the distribution of sample means will approach a normal distribution, regardless of the original population's distribution, as the sample size increases.
This means that even if the original data is not normally distributed, the distribution of sample means will tends towards a bell-shaped curve when you take sufficiently large samples.

15. What is the application of Z statistics in hypothesis testing?
- Z-statistics is a type of hypothesis test used to determine whether a sample mean significantly differ from a known population mean - especially when the population standard deviation is known and the sample size is large(n >= 30).
Applications in Hypothesis Testing are:
1) One-Sample Z-test: Compare sample mean to a known population mean.
2) Two-Sample Z-test: Compare means of two independent large samples.
3) Z-test for proportions: Test whether a sample proportion differs from a population proportion.
4) Confidence Intervals: Build intervals to estimate population parameters.
5) Quality Control and Manufacturing: Test if the mean output deviates from the standard.

16. How do you calculate a Z-score, and what does it represent?
- A z-score, also known as standard score, indicates how many standard deviations a data point is away from the mean of a dataset. It is calculated by subtracting data points with the population mean and then dividing by the standard deviation.
    A Z-score tells you how many standard deviations a data point is from the mean of a distribution.
    Like: if Z = 0, the value is exactly at the mean.
    if Z = +1, one standard deviation above the mean.
    if Z = -2, two standard deviation below the mean.

17. What are point estimates and interval estimates in statistics?
- A point estimate gives a single value as an estimate of a population parameter while an interval estimate gives a range of values, along with full confidence level, within which the parameter is likely to fall.

18. What is the significance of confidence intervals in statistical analysis?
- Significance:
i) A CI reflects how confident we are that our estimate contains the true population parameter.
ii) A CI gives both an estimates and margin of error, showing precision.
iii) CI helps in assessing risk and reliability.
iv) CI quantifies risk in predictions.

19. What is the relationship between a Z-score and a confidence interval?
- The Z-score is used to calculate the margin of error within a confidence interval, which defines the upper and lower bounds of the interval.

20. How are Z-scores used to compare different distributions?
- Z-scores standardize data by expressing values in terms of how many standard deviations they are from the mean. This allows you to compare data points from different distribution - even if they have different units, scales, or spreads.

21. What are the assumptions for applying the Central Limit Theorem?
- Assumptions of CLT are:
i) The sampled values must be independent - the value of one observation should not influence another.
ii) The sample must be drawn randomly from the population to ensure unbiased estimation.
iii) The population should have a finite mean and finite variance.
iv) Sample size should be sufficiently large.
v) Each observation must come from the same distribution.

22. What is the concept of expected value in a probability distribution?
- The expected value of a random variable is the weighted average of all possible values it can take, weighted by their probabilities.

23. How does a probability distribution relate to the expected outcome of a random variable?
- A probability distribution defines the likelihood of all possible outcomes for a random variable, and the expected value is a way to summarize the "average" outcome based on that distribution.
Essentially, the probability distribution tells you how likely each outcome is, and the expected value uses that information to calculate a weighted average of those outcomes.

```python
#18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
def calculate_z_score(x, mean, std):

    z = (x - mean) / std
    return z

x = 115
mean = 100
std = 15

z_score = calculate_z_score(x, mean, std)
print(f"Z-score for x = {x} is: {z_score:.2f}")

# Plot Standard Normal Distribution
x_vals = np.linspace(-4, 4, 1000)
y_vals = norm.pdf(x_vals)

plt.figure(figsize=(10, 4))
plt.plot(x_vals, y_vals, color='skyblue', lw=2, label='Standard Normal PDF')

plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
plt.fill_between(x_vals, y_vals, where=(x_vals <= z_score), color='lightcoral', alpha=0.5)

plt.title('Standard Normal Distribution and Z-score')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#19. Implement hypothesis testing using Z-statistics for a sample dataset.
#Sample Data
sample_mean = 48500
population_mean = 50000
population_std = 6000
sample_size = 100
alpha = 0.05

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
print(f"Z-score = {z_score:.2f}")

z_critical = norm.ppf(1 - alpha / 2)
print(f"Z-critical () = {z_critical:.2f}")

if abs(z_score) > z_critical:
    print("Reject the null hypothesis (H0): There is a significant difference.")
else:
    print("Fail to reject the null hypothesis (H0): No significant difference.")
```

```python
#20. Create a confidence interval for a dataset using Python and interpret the result.

data = np.random.normal(loc=75, scale=10, size=100)  # mean=75, std=10, n=100

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
confidence = 0.95
alpha = 1 - confidence


z_critical = norm.ppf(1 - alpha / 2)

margin_of_error = z_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

def normal_pdf(x, mean=0, std=1):
    coef = 1 / (std * np.sqrt(2 * np.pi))
    exponent = np.exp(-0.5 * ((x - mean) / std)**2)
    return coef * exponent

mean = 50
std_dev = 10
x_vals = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

pdf_vals = normal_pdf(x_vals, mean=mean, std=std_dev)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x_vals, pdf_vals, color='darkcyan', lw=2, label=f'N({mean}, {std_dev})')
plt.title('Probability Density Function of Normal Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.legend()
plt.show()
```

```python
#23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

lambda_val = 4
x_vals = np.arange(0, 15)

cdf_vals = poisson.cdf(x_vals, mu=lambda_val)

for x, cdf in zip(x_vals, cdf_vals):
    print(f"P(X  {x}) = {cdf:.4f}")

# Plot the CDF
plt.figure(figsize=(8, 4))
plt.step(x_vals, cdf_vals, where='post', color='teal', label=f'Poisson(={lambda_val}) CDF')
plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution')
plt.xlabel('Number of Events (x)')
plt.ylabel('Cumulative Probability')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skew_kurtosis(size=1000, mean=0, std=1, seed=42):
    np.random.seed(seed)
    data = np.random.normal(loc=mean, scale=std, size=size)

    skewness_val = skew(data)
    kurtosis_val = kurtosis(data)

    print(f"Skewness: {skewness_val:.4f}")
    print(f"Kurtosis (excess): {kurtosis_val:.4f}")

calculate_skew_kurtosis()
```
